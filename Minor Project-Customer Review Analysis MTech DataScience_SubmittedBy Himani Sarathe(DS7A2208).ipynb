# getting access to a site

```python
import requests
```

# Extracting reviews

```python
from bs4 import BeautifulSoup as bs
```

# For storing final reviews

```python
reviews=[]
```

# #extracting reviews from first 10 pages

```python
for i in range(1,11):
    #extracting reviews from first 10 pages
    page=[]
    #store reviews in a single page
    response=requests.get(url+str(i))
    soup=bs(response.content,'html.parser')
    # creating soup object to iterate over the extracted content 
    rev=soup.findAll('div',attrs={'class',''})
    # Extracting the content under specific tags 
    #tag is 'div' and class is ''(NULL)
    #to know the tag, click on the review-->right click-->select inspect
    for i in range(len(rev)):
        page.append(rev[i].text)
    reviews=reviews+page
```

# writng reviews in a text file 

```python
with open("D:\Mtech Datascience\Projects_ML_mtech\myreview.txt","w",encoding='utf8') as output:#w- wrie r-read
    for i in reviews:
        output.write(i+"\n\n")
```

# Removing unwanted characters from reviews using re function

```python
import re
reviews=" ".join(reviews)
reviews
```

# converting list into a single string because re only work in string

```python
reviews=re.sub('[^A-Za-z' ']+',' ',reviews).lower()
reviews
```

# converting the string to list to remove unwanted words

```python
review_words=reviews.split()
from nltk.corpus import stopwords
sw=stopwords.words('english')
sw
```

# removing the words in stop words

```python
review_words=[i for i in review_words if i not in sw]
```

# we can plot wordcloud for better understanding
#wordcloud only accept string values so again convert list to str

```python
reviews=" ".join(review_words)
reviews
from wordcloud import WordCloud
plot=WordCloud().generate(reviews)
```

```python
import matplotlib.pyplot as plt
plt.imshow(plot)
```

# Extracting Positive Reviews From The Wordcloud

```python
with open("D:\\positive.txt",'r') as  op:
    pos_word=op.read().split()
pos_word
```

```python
pos_words=[i for i in review_words if i in pos_word]
pos_words=" ".join(pos_words)
pos_wordcloud=WordCloud().generate(pos_words)
plt.imshow(pos_wordcloud)
```

# Filtering negative words

```python

neg_words=[i for i in review_words if i not in pos_word]
neg_words=" ".join(neg_words)
neg_words
neg_wordcloud=WordCloud().generate(neg_words)
plt.imshow(neg_wordcloud)
```
